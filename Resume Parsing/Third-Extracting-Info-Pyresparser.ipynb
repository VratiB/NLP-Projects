## Using Pyresparser

```python
import os
os.chdir(dir_Path)
file_List = os.listdir()
```

```python
len(file_List)
```

```python
l2_skills=[]
l2_email=[]
l2_mob=[]
```

```python
for file_Name in file_List:
  print(ResumeParser(file_Name).get_extracted_data())
```

```python
for file_Name in file_List:
  d1=(ResumeParser(file_Name).get_extracted_data())

  skills=d1['skills']
  s=','.join(skills)
  email=d1['email']
  mob=d1['mobile_number']

  l2_skills.append(s)
  l2_email.append(email)
  l2_mob.append(mob)
```

```python
#docx + pdf
l_comb_skills=l1_skills + l2_skills
l_comb_email=l1_email + l2_email
l_comb_mob=l1_mob + l2_mob
```

```python
l_user=[i for i in range(len(l_comb_skills))]
```

```python
import pandas as pd
```

```python
df_1=pd.DataFrame({'Skills':l_comb_skills,'Email':l_comb_email,'Mobile No': l_comb_mob,'UserID':l_user})
```

```python
df_1['Mobile No'] = df_1['Mobile No'].astype(str)
```
