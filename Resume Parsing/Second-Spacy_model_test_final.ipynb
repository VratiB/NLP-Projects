## For Docx Files

```python
doc_type=input('Please enter type of document. If .pdf enter PDF. If .docx enter DOCX: ')
```

```python
if doc_type == 'PDF':
  dir_Path = '/content/drive/MyDrive/NLP-Presentation/PDF-Resume-Test'

if doc_type == 'DOCX':
  dir_Path = '/content/drive/MyDrive/NLP-Presentation/DOCX-Resume-Test'

```

```python
l_docx=[]
if doc_type == 'DOCX':
  os.chdir(dir_Path)
  file_List = os.listdir()
  for file_Name in file_List:
    document = Document(file_Name)
    l_join=[]
    for p in document.paragraphs:
      l_join.append(p.text)
    res = " ".join([str(item) for item in l_join])
    l_docx.append(res)
      
```

```python
len(l_docx)
```

## Cleaning text

```python
 import re
```

```python
docx=[]

for i in l_docx:
  r1=re.findall('[a-zA-Z0-9]+',i)
  r2=' '.join(r1)
  docx.append(r2)
```

```python
docx[0]
```

## Extracting Textual Data from .pdf Files

```python
doc_type=input('Please enter type of document. If .pdf enter PDF. If .docx enter DOCX: ')
```

```python
if doc_type == 'PDF':
  dir_Path = '/content/drive/MyDrive/NLP-Presentation/PDF-Resume-Test'

if doc_type == 'DOCX':
  dir_Path = '/content/drive/MyDrive/NLP-Presentation/DOCX-Resume-Test'

```

```python
l_pdf=[]
if doc_type == 'PDF':
  os.chdir(dir_Path)
  file_List = os.listdir()
  for file_Name in file_List:
    pdfFileObj = open(file_Name,'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    pageObj = pdfReader.getPage(0) 
    l_pdf.append(pageObj.extractText()) 

```

```python
pdf=[]

for i in l_pdf:
  r1=re.findall('[a-zA-Z0-9]+',i)
  r2=' '.join(r1)
  pdf.append(r2)
```

```python
pdf[0]
```

```python
len(pdf)
```

## Now we have a clean textual data in the lists

```python
l_comb=docx + pdf
```

```python
s_name={}
```

```python
count=0
for i in l_comb:
  nlp_doc=model(i)

  for ent in nlp_doc.ents:
    if ent.label_ == 'Name':
      s_name.update({count:ent.text})
      break

    if ent.label != 'Name':
      s_name.update({count:'NameNotFound'})

  count=count+1
```
